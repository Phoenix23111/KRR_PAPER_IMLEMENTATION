```python
BASE_PATH = "MELD.Raw"

TRAIN_VIDEO_PATH = f"{BASE_PATH}/train_splits"
DEV_VIDEO_PATH   = f"{BASE_PATH}/dev_splits"
TEST_VIDEO_PATH  = f"{BASE_PATH}/test_splits/output_repeated_splits_test"

TRAIN_CSV = f"{BASE_PATH}/train_sent_emo.csv"
DEV_CSV   = f"{BASE_PATH}/dev_sent_emo.csv"
TEST_CSV  = f"{BASE_PATH}/test_sent_emo.csv"
```

```python
import pandas as pd

train_df = pd.read_csv(TRAIN_CSV)
dev_df   = pd.read_csv(DEV_CSV)
test_df  = pd.read_csv(TEST_CSV)

print(train_df.columns)
```

```python
import os
import subprocess
from tqdm import tqdm

def extract_audio(video_dir, audio_out):
    os.makedirs(audio_out, exist_ok=True)
    for file in tqdm(os.listdir(video_dir)):
        if file.endswith(".mp4"):
            video_path = os.path.join(video_dir, file)
            audio_path = os.path.join(audio_out, file.replace(".mp4", ".wav"))
            subprocess.call([
                "ffmpeg", "-loglevel", "quiet",
                "-i", video_path,
                "-ac", "1", "-ar", "16000", audio_path
            ])

extract_audio(TRAIN_VIDEO_PATH, "train_audio")
extract_audio(DEV_VIDEO_PATH,   "dev_audio")
extract_audio(TEST_VIDEO_PATH,  "test_audio")
```

```python
import librosa
import numpy as np

def extract_mfcc(audio_folder):
    features = {}
    for file in os.listdir(audio_folder):
        if file.endswith(".wav"):
            y, sr = librosa.load(os.path.join(audio_folder, file), sr=16000)

            mfcc = librosa.feature.mfcc(
                y=y,
                sr=sr,
                n_mfcc=40,
                n_fft=512,
                hop_length=256
            )

            features[file.replace(".wav","")] = np.mean(mfcc, axis=1)
    return features


train_audio_feat = extract_mfcc("train_audio")
dev_audio_feat   = extract_mfcc("dev_audio")
test_audio_feat  = extract_mfcc("test_audio")
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer(max_features=3000)

X_text_train = vectorizer.fit_transform(train_df["Utterance"]).toarray()
X_text_dev   = vectorizer.transform(dev_df["Utterance"]).toarray()
X_text_test  = vectorizer.transform(test_df["Utterance"]).toarray()
```

```python
def align_audio(df, audio_dict):
    audio_features = []
    for _, row in df.iterrows():
        key = f"dia{row.Dialogue_ID}_utt{row.Utterance_ID}"
        audio_features.append(audio_dict.get(key, np.zeros(40)))
    return np.array(audio_features)

X_audio_train = align_audio(train_df, train_audio_feat)
X_audio_dev   = align_audio(dev_df, dev_audio_feat)
X_audio_test  = align_audio(test_df, test_audio_feat)
```

```python
X_train = np.concatenate([X_text_train, X_audio_train], axis=1)
X_dev   = np.concatenate([X_text_dev, X_audio_dev], axis=1)
X_test  = np.concatenate([X_text_test, X_audio_test], axis=1)

y_train = train_df["Emotion"].astype("category").cat.codes
y_dev   = dev_df["Emotion"].astype("category").cat.codes
y_test  = test_df["Emotion"].astype("category").cat.codes
```

```python
from sklearn.linear_model import LogisticRegression

clf = LogisticRegression(max_iter=1000)
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import accuracy_score, f1_score, classification_report, confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

y_pred = clf.predict(X_test)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Macro F1:", f1_score(y_test, y_pred, average="macro"))
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=False, cmap="Blues")
plt.title("Confusion Matrix – MELD Baseline")
plt.show()
```

```python
import tensorflow as tf
from tensorflow.keras.layers import (
    Input, Dense, LSTM, Bidirectional, Dropout,
    Concatenate, Layer
)
from tensorflow.keras.models import Model
from sklearn.utils.class_weight import compute_class_weight
import numpy as np
```

```python
class AttentionLayer(Layer):
    def __init__(self):
        super().__init__()

    def call(self, inputs):
        score = tf.nn.tanh(inputs)
        attention_weights = tf.nn.softmax(score, axis=1)
        context = attention_weights * inputs
        return tf.reduce_sum(context, axis=1)
```

```python
def build_proposed_model(
    text_shape,
    audio_shape,
    num_classes
):
    # ---- Text branch ----
    text_input = Input(shape=text_shape, name="text_input")
    text_lstm = Bidirectional(
        LSTM(64, return_sequences=True)
    )(text_input)
    text_attn = AttentionLayer()(text_lstm)

    # ---- Audio branch ----
    audio_input = Input(shape=audio_shape, name="audio_input")
    audio_lstm = Bidirectional(
        LSTM(64, return_sequences=True)
    )(audio_input)
    audio_attn = AttentionLayer()(audio_lstm)

    # ---- Fusion ----
    fused = Concatenate()([text_attn, audio_attn])
    fused = Dense(128, activation="relu")(fused)
    fused = Dropout(0.4)(fused)

    output = Dense(num_classes, activation="softmax")(fused)

    model = Model(
        inputs=[text_input, audio_input],
        outputs=output
    )

    return model
```

```python
num_classes = len(np.unique(y_train))

model = build_proposed_model(
    text_shape=X_text_train.shape[1:],
    audio_shape=X_audio_train.shape[1:],
    num_classes=num_classes
)

model.compile(
    optimizer="adam",
    loss="sparse_categorical_crossentropy",
    metrics=["accuracy"]
)

model.summary()
```

```python
class_weights = compute_class_weight(
    class_weight="balanced",
    classes=np.unique(y_train),
    y=y_train
)

class_weights = dict(enumerate(class_weights))
print("Class weights:", class_weights)
```

```python
history = model.fit(
    [X_text_train, X_audio_train],
    y_train,
    validation_split=0.1,
    epochs=15,
    batch_size=32,
    class_weight=class_weights,
    verbose=1
)
```

```python
from sklearn.metrics import accuracy_score, f1_score, classification_report

y_pred = model.predict([X_text_test, X_audio_test])
y_pred = np.argmax(y_pred, axis=1)

print("Accuracy:", accuracy_score(y_test, y_pred))
print("Macro F1:", f1_score(y_test, y_pred, average="macro"))
print(classification_report(y_test, y_pred))
```
